```python
from loguru import logger
import torch
def analyze_patch_distribution(patch, thresholds=[0.0, 0.1, 0.3, 0.5]):
    value = patch['INTERPOLATED_ANA']
    total_voxels = value.numel()
        
    logger.info(f"\n=== Single Patch Analysis ===")
    logger.info(f"value shape: {value.shape}")
    logger.info(f"value range: [{value.min():.4f}, {value.max():.4f}]")
        
    for t in thresholds:
        positive_voxels = (value > t).sum().item()
        percentage = (positive_voxels / total_voxels) * 100
        logger.info(f"Threshold {t:.1f}:")
        logger.info(f"- Positive voxels: {positive_voxels}/{total_voxels} ({percentage:.2f}%)")

def _is_positive_patch(patch, threshold=0.5, min_ratio=0.05):
    value = patch['INTERPOLATED_ANA']
    positive_ratio = (value > threshold).sum() / value.numel()
    return positive_ratio > min_ratio   
all_patches = torch.load('/Users/keyi/Desktop/ds_11_17_multilabel_4tries/per_sample_1_train/sample_0_train.pt')

for patch in all_patches:
    analyze_patch_distribution(patch)
patch_stats = []
for patch in all_patches:
    value = patch['INTERPOLATED_ANA']
    pos_ratio = (value > 0.5).sum() / value.numel()
    patch_stats.append(pos_ratio.item())
logger.info("\n=== Overall Statistics ===")
logger.info(f"Patches with >1% positive voxels: {sum(r > 0.01 for r in patch_stats)}")
logger.info(f"Patches with >5% positive voxels: {sum(r > 0.05 for r in patch_stats)}")
logger.info(f"Patches with >10% positive voxels: {sum(r > 0.1 for r in patch_stats)}")

```
