```python
import json
for index in range(5):
    with open(f'/Users/keyi/Desktop/DL_template/_assets/split_json/AutoPET_Cluster_Dev/dataset_split_{index}.json', 'r') as f:
        data = json.load(f)
    for entry in data['TRAINING']:
        ctres_path = entry['CTRES']
        interpolated_ana_path = ctres_path.replace('CTres.nii.gz', 'interpolated_ANASEG.nii.gz')
        entry['INTERPOLATED_ANA'] = interpolated_ana_path
    for entry in data['VALIDATION']:
        ctres_path = entry['CTRES']
        interpolated_ana_path = ctres_path.replace('CTres.nii.gz', 'interpolated_ANASEG.nii.gz')
        entry['INTERPOLATED_ANA'] = interpolated_ana_path
    print(data)
    with open(f'/Users/keyi/Desktop/DL_template/_assets/split_json/AutoPET_Cluster_Dev/dataset_split_{index}.json', 'w') as f:
        json.dump(data, f,indent=4)
```
